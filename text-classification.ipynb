```python
category_mapping = {
    'Documentary':'Documentary',
    'Documentary-Music':'Documentary',
    'Music-Documentary':'Documentary',
    'Documentary-History':'Documentary',
    'Music':'Documentary',
    'Drama':'Drama',
    'Drama-Romance':'Drama',
    'Comedy-Drama':'Comedy',
    'Drama-Comedy':'Drama',
    'Romance-Drama':'Romance',
    'Drama-TV Movie':'Drama',
    'Crime-Drama':'Drama',
    'Drama-Crime':'Drama',
    'Drama-Family':'Drama',
    'Drama-History':'Drama',
    'Comedy':'Comedy',
    'Comedy-Romance':'Comedy',
    'Horror':'Horror-Thriller',
    'Horror-Thriller':'Horror-Thriller',
    'Thriller':'Horror-Thriller',
    'Action':'Action',
    'Action-Drama':'Action',
    'Romance':'Romance',
    'Romance-Comedy':'Romance',
    'Animation':'Fiction',
    'Science Fiction':'Fiction',
    'Western':'Action',
    'Family':'Drama',
    'Animation-Family':'Fiction',
    'Fantasy':'Fiction',
    'Animation-Comedy':'Fiction',
    'Horror-Comedy':'Comedy',
    'Thriller-Horror':'Horror-Thriller'

}
```

```python
movies["simple_categories"]=movies["genres"].map(category_mapping)
```

```python
movies[~(movies["simple_categories"].isna())]
```

```python

from transformers import pipeline

categories2=["Documentary","Drama"]
pipe=pipeline("zero-shot-classification",model="facebook/bart-large-mnli",device="mps")
```

```python
sequence=movies.loc[movies["simple_categories"]=="Drama","overview"].reset_index(drop=True)[0]
```

```python
pipe(sequence,categories2)
```

```python
import numpy as np
max_index=np.argmax(pipe(sequence,categories2)["scores"])
max_label=pipe(sequence,categories2)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence,categories):
    predictions=pipe(sequence,categories)
    max_index=np.argmax(predictions["scores"])
    max_label=predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats=[]
predicted_cats=[]

for i in tqdm(range(0,300)):
    sequence=movies.loc[movies["simple_categories"]=="Drama","overview"].reset_index(drop=True)[i]
    predicted_cats+=[generate_predictions(sequence,categories2)]
    actual_cats+=["Drama"]
```

```python
for i in tqdm(range(0,300)):
    sequence=movies.loc[movies["simple_categories"]=="Documentary","overview"].reset_index(drop=True)[i]
    predicted_cats+=[generate_predictions(sequence,categories2)]
    actual_cats+=["Documentary"]
```

```python
predictions_df=pd.DataFrame({"actual_categories":actual_cats,"predicted_categories":predicted_cats})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"]=(
    np.where(predictions_df["actual_categories"]==predictions_df["predicted_categories"],1,0)
)
```

```python
predictions_df["correct_prediction"].sum()/len(predictions_df)
```

```python
ids=[]
predicted_cats=[]

missing_cats=movies.loc[movies["simple_categories"].isna(),["id","overview"]].reset_index(drop=True)
```

```python
missing_cats
```

```python
for i in tqdm(range(0,len(missing_cats))):
    sequence=missing_cats["overview"][i]
    predicted_cats+=[generate_predictions(sequence,categories2)]
    ids+=[missing_cats["id"][i]]
```

```python
missing_predicted_df=pd.DataFrame({"id":ids,"predicted_categories":predicted_cats})
```

```python
missing_predicted_df
```

```python
movies=pd.merge(movies,missing_predicted_df,on="id",how="left")
```

```python
movies
```

```python
movies["simple_categories"]=np.where(movies["simple_categories"].isna(),movies["predicted_categories"],movies["simple_categories"])
movies=movies.drop(columns=["predicted_categories"])
```
